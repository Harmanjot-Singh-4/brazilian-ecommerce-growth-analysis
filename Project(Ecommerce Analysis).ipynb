<h1>Insights from this Datasets
<h1>

<h4>1. From Which location the most customers are coming from ?<h4>

```python
top_cities = customers['customer_city'].value_counts().head(10)
print('Most Customers are from these Cities:',top_cities)
print()
top_states = customers['customer_state'].value_counts()
print('Most Customers are from these States:',top_states)
```

<h4>2. Highest selling and Lowest selling Product Categories ?<h4>


```python
products_eng = pd.merge(products, products_tra, on='product_category_name', how='left')

sales_data = pd.merge(order_items, products_eng, on='product_id', how='inner')

highest_revenue = sales_data.groupby('product_category_name_english')['price'].sum().sort_values(ascending=False).head(10)
print('Highest Selling Product Categories:',highest_revenue)
print()
lowest_revenue = sales_data.groupby('product_category_name_english')['price'].sum().sort_values(ascending=True).head(10)
print('Lowest Selling Product Categories:',lowest_revenue)
```

<h4>3. Which Brazilian State has the highest gap between the Estimated Delivery Date and the Actual Delivery Date? <h4>

```python
orders['order_delay_delivery_days'] = (orders['order_delivered_customer_date'] - orders['order_estimated_delivery_date']).dt.days

location_delivery = pd.merge(customers, orders, on='customer_id', how='inner')

delivery_delays = location_delivery.groupby(['customer_state'])['order_delay_delivery_days'].mean().sort_values(ascending=True)

print(delivery_delays)
```

<h4>4. Are customers coming back? What is the percentage of customers who have placed more than one order?<h4>

```python
customer_loyalty = customers.groupby('customer_unique_id').size()

repeat_customers = (customer_loyalty > 1).sum()

total_unique = customers['customer_unique_id'].nunique()

retention_rate = (repeat_customers / total_unique) * 100

print(f"Total Unique Customers: {total_unique}")
print(f"Repeat Customers: {repeat_customers}")
print(f"Retention Rate: {retention_rate:.2f}%")
```

```python
plt.figure(figsize=(8,8))

labels = ['One-time Customers', 'Repeat Customers']
sizes = [total_unique - repeat_customers, repeat_customers] 
plt.pie(sizes, labels=labels, autopct='%1.2f%%', startangle=140, colors=['lightgrey', 'gold'], explode=(0, 0.1))
plt.title('Customer Retention Rate')
plt.savefig('retention_pie.png')
```
